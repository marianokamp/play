```python
image_downloads = [('formal_attire.txt', 'images/formal'), ('sports_attire.txt', 'images/sports')]
for (urls, dest) in image_downloads:
    download_images(urls=urls, dest=dest, max_workers=5, max_pics=5000)
    verify_images(dest, delete=True, max_size=500)
```

```python
np.random.seed(42)
data = ImageDataBunch.from_folder('images', train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4, bs=128).normalize(imagenet_stats)
```

```python
(data.classes, data.c, len(data.train_ds), len(data.valid_ds))
```

```python
data.show_batch(rows=3, figsize=(7,8))
```

# Learning

```python
from fastai.metrics import error_rate
```

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```python
#learn.model
```

```python
%%time
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

```python
learn.load('stage-1')
```

```python
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
interp.plot_top_losses(9, figsize=(15,11))
```

```python
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```python
interp.most_confused(min_val=3)
```
